```python
import numpy as np
import pandas as pd 
from bs4 import BeautifulSoup
import requests
from datetime import date
import os
```

```python
def scrap_thenumbers_worldwide(year):
    url=f'https://www.the-numbers.com/movies/year/{year}'
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
    }


    req=requests.get(url, headers=headers)
    content=req.text
    soup=BeautifulSoup(content)
    appended_data = []
    rows = soup.find_all('tr')[1:]
    os.makedirs('datatheNumbers', exist_ok=True)
    for index in range(0,len(rows)):
        data_row = {}
        data = rows[index].findAll('td')
        if len(data) <=1:
            continue
        if len(data)==6 :
            data_row['Movie'] = data[1].text
            data_row['Genre'] = data[2].text
            data_row['Release Type'] = data[3].text
            data_row['Revenue to Date'] = data[4].text
            data_row['Year'] = year
        if len(data)==5:
            data_row['Movie'] = data[0].text
            data_row['Genre'] = data[1].text
            data_row['Release Type'] = data[2].text
            data_row['Revenue to Date'] = data[3].text
            data_row['Year'] = year
        appended_data.append(data_row)
    ranking_data = pd.DataFrame(appended_data, columns = ['Movie', 'Genre', 'Release Type', 'Revenue to Date' ,'Year']) 
    ranking_data.to_csv( f'datatheNumbers/ranking_summary_{year}.csv',index=False)


```

```python
todays_date = date.today()
current_year = todays_date.year
years = range(1899, current_year+1)
```

```python
for year in years:
    print(year)
    scrap_thenumbers_worldwide(year)
```
